```python
data = [
    {"month": "Jan", "region": "North", "sales": 1200},
    {"month": "Jan", "region": "South", "sales": 900},
    {"month": "Feb", "region": "North", "sales": 1400},
    {"month": "Feb", "region": "South", "sales": 1100},
    {"month": "Mar", "region": "North", "sales": 1000},
    {"month": "Mar", "region": "South", "sales": 1300},
]

total_sales = 0

for row in data:
    total_sales += row["sales"]

print (f"Total Sales : {total_sales}")
```

```python
sales_by_region = {}

for row in data:
    region = row["region"]
    sales = row["sales"]

    if region not in sales_by_region:
        sales_by_region[region] = 0

    sales_by_region[region] += sales

print(sales_by_region)
```

```python
sales_by_month = {}

for row in data:
    month = row["month"]
    sales = row["sales"]

    if month not in sales_by_month:
        sales_by_month[month] = 0
    
    sales_by_month[month] += sales

print(sales_by_month)

best_month = max(sales_by_month,key=sales_by_month.get)

print(best_month)
print(f"Best Month: {best_month}")
print(f"Amount Made: {sales_by_month[best_month]}")
```
